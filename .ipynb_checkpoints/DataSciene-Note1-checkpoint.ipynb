# Feature Transformiaton
- Some Machine Learning models, like Linear and Logistic regression, assume that the variables follow a normal distribution. More likely, variables in real datasets will follow a skewed distribution.
- By applying some transformations to these skewed variables, we can map this skewed distribution to a normal distribution so, this can increase the performance of our models.

Sklearn has three Transformations-

1.  Function Transformation

2.  Power Transformation

3.  Quantile transformation

## Function Transformation
### Log Transformation
   1. bu dönüşüm ile verimiz normale akın bir dağılım gösterir
   2. genellikle sağa çarpık verilerde kullanılır (right skewed data)
   3. 0 veya negatif değere sahip verilerde kullanılamaz bunun için log(x+1) np.log1p() kullanılabilir
   4. $ x -> \log(x) $ 
### Reciprocal Transformation
   1. 0 da tanımlı değildir
   2. It is a powerful transformation with a radical effect.
   3. This transformation reverses the order among values of the same sign, so large values become smaller and vice-versa.
   4. $ x -> 1/x $

### Square Transformation
   1. genellikle sola çarpık data için kullanılır.
### Square root transformation
   1. sadece positif sayılar için tanımlıdır
   2. log transformationun daha zayıf halidir
   3. sağa çarpık verilerin çarpıklıgını azalatmak için kullanılabilir
   4. $ x -> \sqrt{x} $ 
## Power Transformations
- İstenilen sonuç "Gaussian" gibi oldugundan kullanılır
- ‘Box-Cox’ and ‘Yeo-Johnson’ transforms.
- 'Box Cox' data pozitif olmak zorunda 0 bile kabul edilmez (bir kaynaga göre doğru en çok kullanılan dönüşüm)
- 'Yeo-Johnson' data negatif ve 0 oldugunda yardımımıza koşar

* *  Yapılan modelin biçimine, rengine, kokusuna göre farklı tipteki dönüşümler kullanılabilir

# Feature Scaling
1. Min Max Scaling
    - It just scales all the data between 0 and 1.
    - 0 ve 1 değerini feature_range ile değiştirebiliriz
    - $x.scaled = (x – x.min)/(x_max – x.min)$
2. Standart Scaling
    - mean is 0 and the standard deviation is 1(or the variance).
    - Standart Scaler dağılımın normal oldugnu varsayar.
    - ya normal dağılıma dönüştürün ya da başka scaler sçein
    - sonuç a bakıldıgında mean 0 a yakın std 1 yakın görülür
    - $x.scaled = x – mean/std.dev $
3. MaxAbsScaler 
    - absolute maximum value of each column and divides each value in the column by the maximum value.
    - [-1,1] arasına ölçekler veriyi
4. Robust Scaler
    - mean,max,min bunlar outilar için çok hassas ama robust değil
    - 1.removes the median from the data 2. scales the data by the (IQR)
    
5. Quantile Transformer Scaler

# Model Hyperparameter with optuna


```python
import optuna

def objective(trial):
    x = trial.suggest_float('x', -10, 10) # x adlı parametreyi -10 dan 10 a dener
    return (x - 2) ** 2                   # sugges_float da türünü belirler her bir paramatre için docstan bak nasıl bir artış 
                                          # return değeri minimize edilmesi istenen değer
study = optuna.create_study(direction ="minimize") 
study.optimize(objective, n_trials=100) #objective devamlı çağrılıcak olan funct

study.best_params  # E.g. {'x': 2.002108042} # sonuçlar

#suggest_discrete_uniform(name, low, high, q)
#Suggest a value for the discrete parameter.
#suggest_float(name, low, high, *[, step, log])
#Suggest a value for the floating point parameter.
# https://optuna.readthedocs.io/en/stable/reference/generated/optuna.trial.Trial.html?highlight=sugges_float#optuna.trial.Trial.suggest_float
```
